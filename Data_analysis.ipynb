```python
import seaborn as sns
sns.countplot(x='CDR',data=df_dementia,color="orange")
```

<b> About 200 people have CDR value 0 that means they are normal </b>

```python
# Categorizing feature CDR
def cat_CDR(n):
    if n == 0:
        return 'Normal'
    
    else:                                        
        return 'Dementia'
```

```python
df_dementia['CDR'] = df_dementia['CDR'].apply(lambda x: cat_CDR(x))
```

```python
plt.figure(figsize=(10,6))
df_dementia['CDR'].value_counts().plot.pie(autopct='%1.2f%%')
```

<b> According to CDR value 55.23% people are normal and rest are suffering from dementia </b>

```python
plt.figure(figsize=(10,7))
sns.countplot(x='MMSE',data=df_dementia,color="grey")
plt.title("Countplot for MMSE Values")
```

<b>MMSE values of 30 has the highest count.</b>

```python
# Categorizing feature MMSE
def cat_MMSE(n):
    if n >= 24:
        return 'Normal'
    elif n <= 9:
        return 'Severe'
    elif n >= 10 and n <= 18:
        return 'Moderate'
    elif n >= 19 and n <= 23:                                   
        return 'Mild'
```

```python
df_dementia['MMSE'] = df_dementia['MMSE'].apply(lambda x: cat_MMSE(x))
```

```python
plt.figure(figsize=(10,6))
df_dementia['MMSE'].value_counts().plot.pie(autopct='%1.2f%%')
```
